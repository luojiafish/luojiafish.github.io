### 池化
>作用：为了在很小的局部范围内做一点点“没那么在乎具体像素位置”的压缩，顺便把特征图变小、计算量降下去、让后面层更容易学。
> - **局部平移不变性**:在一个小窗口里（比如 2×2），不太在乎特征精确落在哪个格子,有强响应就行，在哪个像素都 OK
> - **增大后面层的有效感受野**
> - **降采样，减少参数和计算量**
> - **降采样，减少参数和计算量**
> - **一定程度的“去噪”和“特征突出”**:max pooling：保留局部里最强的响应，弱响应被压掉；avg pooling：平滑局部响应，去掉一点点高频噪声。

H_{\text{out}} = \left\lfloor \frac{H_{in} + 2p_h - k_h}{s_h} \right\rfloor + 1
$$

W_{\text{out}} = \left\lfloor \frac{W_{in} + 2p_w - k_w}{s_w} \right\rfloor + 1
