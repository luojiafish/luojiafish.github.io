# 目标
- 直流电动机转矩如何产生
- 交流电动机转矩如何产生
- 了解矢量控制，就是将所有电动机等效为他励直流电机，一套绕柱（一个电流分量）控制电机的励磁，另一套绕柱（另一个电流分量）控制电动机的转矩，从而让交流变成直流来控制


## 基础知识
与磁场相关的物理量,机及相关定律

磁场强度：$H$

磁感应强度：
$
\vec{B} = \mu \vec{H} = \mu_r \mu_0 \vec{H}
$

磁通量（$\Phi$读作菲）：
$
\Phi = \vec{B} \cdot \vec{S}
$

磁导率（$\mu_r$相对磁导率和$\mu_0$真空磁导率）：
$
\mu = \mu_r \mu_0
$

磁阻：$R = \frac{l}{\mu S}$

磁导：$\Lambda = \frac{1}{R}$

磁动势：$f_{\mathrm{A}}=N_{A}i_{A}$ 

励磁磁链定义（$\psi$读作赛，磁通量与匝数的乘积，由于磁路是串联的，磁链铁心与气隙相等）：
$\psi_{A} = \phi_{A} N_A$

线圈A的励磁电感：$L_{A} =  N_A^2 \frac{\Lambda_m \Lambda_\delta}{\Lambda_m + \Lambda_\delta} = L_{mA} + L_{\sigma A}(称为漏感)$，
其中 $L_{\sigma A}$ 表示漏电感，主要分布在空气中，近似为常量，由漏电感是衡量漏励磁，漏磁链就是指主要在绕组A附近闭合、没进入主磁路（即铁心）、对其他绕组链接很弱的那部分，在有气隙的系统里，“是否穿过气隙”就成了区分主、漏磁链的标准

安培环路定律（如 [图1](#安培环路定律)）：$$\oint_{L} \boldsymbol{H} \cdot \mathrm{d}\boldsymbol{l} = \sum i \tag{1-1}$$

## 单线圈静止的绕磁
（忽略绕组B，设铁芯长度为 $l_m$，气隙长度为 $\delta$，铁芯内的磁场强度为 $\vec{H}_m$，铁芯内的磁通密度（或磁感应强度）为 $B_m$，隙内的磁感应强度为 $\vec{B}_\delta$，气隙内的磁场强度为 $\vec{H}_\delta$，且铁芯截面积与气隙截面积相等，均为 $S$；不考虑气隙磁场畸变，则铁芯内的磁通量与气隙中的磁通量相等，即 $\Phi_m = \Phi_\delta = \Phi$。

对于[图2](#安培环路定律)，如果忽略B线圈，那么有$$H_{m}l_{m}+H_{\delta}l_{\delta} = N_{A}i_{A}\tag{1-2}$$
图中虚线为闭合环路，此时线圈A相当于产生磁场$H$的“磁场源”，可以想到电路中的电源

对式 (1-2)推导后得到：$$f_{\mathrm{A}} = \frac{B_{\mathrm{m}}}{\mu_{\mathrm{Fe}}} l_{\mathrm{m}} + \frac{B_{\delta}}{\mu_0} \delta = \left(B_{\mathrm{m}} S\right)\left( \frac{l_{\mathrm{m}}}{\mu_{\mathrm{Fe}} S} \right) + \left(B_{\delta} S\right)\left( \frac{\delta}{\mu_0 S} \right) = \phi_{\mathrm{mA}} R_{\mathrm{m}} + \phi_{\delta} R_{\delta} = \phi_{A} R_{A}\tag{1-4}$$

也就是把带电线圈看作电压，磁路总磁阻看作电阻，磁通量看作电流，式（1-4）称为磁路中的欧姆定律

由于气隙磁阻$>>$铁心磁阻，故忽略铁心，推导得：
$$
\psi_{A}
= N_A \Phi_A
= N_A f_A \frac{\Lambda_m \Lambda_\delta}{\Lambda_m + \Lambda_\delta}
= N_A^2 i_A \frac{\Lambda_m \Lambda_\delta}{\Lambda_m + \Lambda_\delta}
$$
其中匝数、整体电导是恒定值，为了衡量线圈产生磁链的能力，故定义励磁电感 $L_{A}$


磁链与电感的联系 （分析得磁链主要由电流决定）：

$$
\psi_{A} = \psi_{\sigma A} + \psi_{mA}
= L_{\sigma A} i_A + L_{mA} i_A
= L_A i_A
$$
对于线圈A，其总磁链就是自感与线圈电流的乘积，由于自感中漏电感近似为常量，最终将电流、匝数和磁阻和线圈产生磁链能力联系起来，同时气隙磁阻$>>$铁心磁阻，铁心上磁压降小，气隙上磁压降大，以又侧重到气隙上





## 磁能
磁链的变化会产生感应电动势$e_{AA}$（表示：由 绕组 A 自身磁链变化 在 绕组 A 中 产生的感应电动势），对于单绕柱系统,总电感等于自感磁链，则$\psi_{AA} = \psi_{A}$，不然 $\psi_{A}$ 还包含互感磁链，规定感应电动势的正方向和电流正方向一样
$$
u_A = i_A R_A + \frac{d \psi_{AA}}{dt}
$$

式子两边乘以绕柱A电流后对时间积分，左边就是电能，右边为电阻损耗（热能）+磁场储能
$$
\int u_A i_A \, dt
=
\int i_A^2 R_A \, dt
+
\int i_A \, d\psi_{AA}
$$
$$
\text{磁能：}\quad
W_m = \int i_A \, d\psi_{AA}
$$
$$
\text{磁共能：}\quad
W_m' = \int \psi_{AA} \, di_A
$$
因为铁心中$B$与$H$是非线性的，推导可知$i$与$\psi$的关系实际就是这个非线性，而由这部分非线性引起的那一部分磁能，确实对应“铁心中的磁场储能” （你就假设假如铁心是线性的，[图4](#磁能与磁共能)会是啥样，不一样在哪），当忽略这部分磁场储能时，磁能等于磁共能

## 双线圈励磁

对于双线圈绕组，相比于单线圈绕组，增加了互感励磁，互感励磁进而影响磁能的计算
已知A线圈的自感为$L_{A}$,互感为$L_{m}$,有
$$ \text{A线圈中总励磁为：}\quad
\psi_{A} = \psi_{AA}+\psi_{AB} = L_{A}i_{A} + L_{m\delta}i_{B} = N_{A}^2 \Lambda_{m\delta} + N_{A}N_{B} \Lambda_{m\delta}$$
对B线圈励磁计算同理可得

$$
\text{A线圈中电压为：}\quad
u_A = i_A R_A + \frac{d \psi_{A}}{dt}
$$
B线圈电压同理，A、B线圈电压公式相加求积分可得：
$$
\text{总磁能：}\quad
W_m = \int i_A \, d\psi_{A} + \int i_B \, d\psi_{A}
$$
同样忽略铁心储能，令N_A = N_B, 有
$$
W_m = W_m' =  \int i_A \, d\psi_{A} + \int i_B \, d\psi_{A} = \int i_A \, d(L_{A}i_{A} + L_{m\delta}i_{B}) + \int i_B \, d(L_{B}i_{B} + L_{m\delta}i_{A}) 
= \tfrac{1}{2}i_A^2 L+\tfrac{1}{2}i_B^2 L+L_{m\delta}i_A i_B
$$

之前都是电 $\leftrightarrow$ 磁，为了有机械能的转化，引入了转动，有如下假定：

- 两套绕组，一套位于定子上，为 $s$；一套位于转子上，为 $r$；
- 绕组匝数相等，$N_s = N_r$；
- 定子与转子气隙处处相等，都为 $g = \sigma / 2$；
- 忽略铁芯内由磁阻引起的磁场储能，主磁路由气隙决定。

对于 $s$ 和 $r$ 的全磁链，同样分为自感和互感。以 $s$ 为例：

自感

$$
\psi_{ss} = \psi_{ms} + \psi_{\sigma s}
= L_{ms} i_s + L_{\sigma s} i_s
= (L_{ms} + L_{\sigma s}) i_s
= L_s i_s
$$

互感
$$
\text{转子产生的主磁通为}\quad
\Phi_m = \Lambda_{m\delta} N_r i_r .
$$

由于转子磁轴与定子绕组磁轴之间存在夹角 $\theta_{sr}$，能够与定子有效耦合的磁通为其在定子磁轴方向上的投影：

$$
\Phi_{sr} = \Phi_m \cos \theta_{sr}.
$$

因此，转子对定子的互磁链为

$$
\psi_{sr}
= N_s \Phi_{sr}
= N_s \Phi_m \cos \theta_{sr}
= N_s N_r \Lambda_{m\delta} i_r \cos \theta_{sr}
= M(\theta_{sr}) i_r,
$$

其中互感定义为

$$
M(\theta_{sr}) = N_s N_r \Lambda_{m\delta} \cos \theta_{sr}.
$$

其中先明确转子磁链方向和 $N/S$ 极性，如 [图5](#机电装置) 所示：当定子电流为图示规定正方向时，在气隙中产生的主磁链由上指向下。根据磁力线在气隙中由 $N \to S$ 的约定，可知上方为 $N$ 极、下方为 $S$ 极。

定、转子之间的互感并非由磁路本身决定，而是由两套绕组磁轴之间的空间夹角决定。由于转子主磁通与定子磁轴之间存在夹角 $\theta_{sr}$，其对定子产生的互磁链等于主磁通在定子磁轴方向上的投影，因此互磁链中自然引入 $\cos \theta_{sr}$ 因子。

$$全磁链s = 自感 + 互感  = \psi_{ss} + \psi_{sr}$$

磁能
$$
W_m = W_m'
= \frac{1}{2} i_s \psi_s + \frac{1}{2} i_r \psi_r
= \frac{1}{2} i_s \bigl( L_s i_s + L_m \cos\theta_{sr}\, i_r \bigr)
+ \frac{1}{2} i_r \bigl( L_m \cos\theta_{sr}\, i_s + L_r i_r \bigr)
= \frac{1}{2} L_s i_s^2
+ L_m \cos \theta_{sr}\, i_s i_r
+ \frac{1}{2} L_r i_r^2
$$



设电磁转矩$T_e$，有：

$$
\text{瞬时机械功率：}\quad
dW_{mech} = T_e d\theta_{rs}
$$

$$
\text{瞬时磁能：}\quad
dW_m
= L_s i_s d i_s
- L_m \sin \theta_{sr}\, i_s i_rd\theta_{sr}
+ L_m \cos \theta_{sr}\, i_rdi_s + L_m \cos \theta_{sr}\, i_sdi_r
+  L_r i_rdi_r 
$$
其中变化量有电流和角度，分别求微分

$$
\text{瞬时输入电能：}\quad
dW_e
= i_r \, d\psi_r + i_s \, d\psi_s
= dW_m + dW_{mech}
$$
其中，式中写成 ($dW_e = id\psi$) 并不意味着假定电流 $i$ 不变。
该关系仅由瞬时功率表达式 $p = ui$ 以及感应电压 $u = d\psi/dt$ 推出，即
$
 p = ui = i \frac{d\psi}{dt} \Rightarrow dW = p,dt = i, d\psi,
$
它只是将输入电能用磁链作为变量来表示，并不对电流随时间变化作任何限制。
实际上，磁链$ \psi $本身是电流和转角的函数，如 $ \psi=\psi(i_r,i_s,\theta_{sr})$，
因而 $d\psi$ 中已自然包含了由 $di_r$、$di_s$及$d\theta_{sr}$ 引起的变化。
只有在后续利用磁能或共能法求电磁转矩时，才会在偏导运算中引入
“保持电流不变”或“保持磁链不变”的条件，而并非在本能量平衡式中作此假定。

对输入电能进一步推导可知：
$$
dW_{mech}
= T_e \, d\theta_{rs}
= - L_m \sin \theta_{sr}\, i_s i_r \, d\theta_{sr}
$$
$$
\text{电磁转矩：}\quad
T_e
=- L_m \sin \theta_{sr} i_s i_r 
=L_m \overrightarrow{i_r} \times \overrightarrow{i_s} 
$$
现在对电磁转矩方向进行判断
