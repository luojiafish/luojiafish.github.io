## 电气传动系统动力学方程
在以转速为方向为基准方向的前提下,规定了电磁转矩和负载转矩的正方向

由旋转运动系统的牛顿第二定律得:

$$
T - T_L = J \frac{d\Omega}{dt}$$


$$
W_m = \frac{1}{2} J \Omega^2
= \frac{G D^2}{2 \cdot 4 g}\,\Omega^2
= \frac{G D^2}{8 g}\,\Omega^2
$$


电气传动系统动力学方程推导如下
$$
T - T_L = \frac{G D^2}{375}\,\frac{dn}{dt}
$$

## 负载转矩和飞轮矩的折算
先搞明白为什么要折算，

1. 电机驱动的过程链或者负载运动情况是负载的，有的还涉及到运动形式的转化
2. 为了控制简单化，就希望知道和电机直接相关的量，等效到电机轴上的负载转矩，飞轮矩
3. 选择合适的电机，以及是否需要额外的增减速器，不然对电机本身要求高，
导致体积大或者工艺成本上升（例如，当负载功率一定，转速下降，转矩升高：电机需要降低速度，根据之前电机学中学到过稳态下：$n = \frac{60 f}{p}$，可知电源频率一定，磁极对数要提高，电机体积增大，降低电源频率则需要驱动器，都会导致工艺复杂，体积增大）

**折算的原则**

负载转矩的折算：折算前后，系统储存的动能相等。对于电动机状态，所有的机械损耗由电动机来提供，电动机输出功率为机械损耗和负载功率之和。$( T_{L折}\Omega_{m折} \eta= T_L\Omega_L )$，对于发电状态,知道负载让电动机动起来，负载提供机械损耗就知道怎么算了，由于电动机，发电机本就是可逆的过程，所以对于相同大小的负载转矩和负载转速，两过程机械损耗的功率大小相等。

飞轮矩：系统所存储动能相等

具体有很多形式，如多轴传动，水平移动等，抓住原则进行等效，原理相同
