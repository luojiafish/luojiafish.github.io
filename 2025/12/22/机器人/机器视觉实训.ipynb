## 检测，主要是利用快速匹配特征，哪怕特征旋转、等比例变换也行，但是仅限于此
### 项目1：工件螺丝孔位检测
使用VisionMaster进行图像检测，快速匹配确定待检测图像，格式化文字显示到图像上，条件检测（看个数或其他图像特征是否符合要求）
### 项目2：密封圈有无检测
### 项目3：按键错漏检测
先选择一个始终固定的特征，然后进行位置修正，然后采用多个快速匹配来分别确定对应按键是否正确，条件判断
```
注意
对于6和9这两按键检测时候角度不能是180°，改到-90°~90°即可
用于定位的特征，匹配分数可以适当的小点，50左右，后接位置修正（先单点单次运行后设置基准），基于定位特征的待检测特征，手绘ROI区域后，选择位置修正信息
待检测特征根据实际情况微调匹配分数

```
### 项目4：多种类型刀头识别检测
快速检测创建特征模板，每种类型创建一个以上，给特征模板改对应的名，然后接格式化输出显示这是XX刀
```
注意
多个特征类型要打开**快速检测**的**全局搜索模式**
```

##  项目5：黑色背景下孔洞数量检测（Blob 分析）
在纯黑的背景下，采用BLOB分析
```
参数
阈值方式：采用单阈值，因为背景是黑的，待检测孔是白的，排除黑的就是白的，所以单阈值
极性：；亮于背景
孔洞最小面积：调解起来没变化，果白孔里面有黑点就忽略黑点
面积使能：规定面积范围筛选出规定的孔洞，有些太小了不符合，就改下限值；如果背景和待检测一个颜色，就改上限值

```
##  项目6：杯壁污渍检测（Blob 分析）
```
难点
数据分析：先把阈值调好，确保待检测特征都出现，把待检测特征和干扰特征数据分好，对每个特征进行分析（最大最小值）
双阈值：调参

```
