# DM-MC-Board02 开发板学习

需掌握
```
STM32H723 最小系统 + 工程跑起来,能下载程序、能跑、能调试断点:
    用 ST-Link / J-Link + SWD 下载程序，能跑一个最小工程（LED / printf / 定时器）
    STM32H7 启动流程（大概）
    CubeMX 基本用法
    时钟树（知道 550MHz 是怎么来的，不用推公式）
CAN,STM32 通过 CAN 给电机发使能 / 位置 / 速度 / 扭矩命令并能接收反馈:
    CAN 引脚（CAN1 / CAN2 / CAN3）
    发送 / 接收流程
电源树 + 可控电源（别炸板）,上电顺序哪个接口能直接插电机:
    12–24V 主电源怎么进板子
    哪些接口是常供电、可控供电（PMOS 控制）
    什么时候能给 CAN / PWM 供电
```
以及定时器，串口



电源树  
电机一个24v电源，电机和开发板用CAN通讯连接，开发板一个12~24v电源
引脚  
BOOT、SWD、CAN 引脚、时钟源、可控电源使能引脚
时钟源：板子有外部 24MHz 晶振；CubeMX：选 HSE = 24MHz，不要选内部 RC可控电源
CAN  
1，2为纯信号接口。哪怕是3号CAN接口连接的设备也需要外部供电，除非是传感器超低功耗产品
