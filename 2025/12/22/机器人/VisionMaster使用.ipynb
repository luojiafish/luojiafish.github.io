## 项目1 OCR识别（Optical Character Recognition，光学字符）+ 数据TCP发送
根据用户手册是最快最科学的方法

方案一：使用快速匹配+位置修正+字符识别+客户端 / 服务器数据接收和发送
```
问题：
设置的ROI区域和显示的不一样：因为没设置，通过快速配置中设置要识别的特征，会给出识别的区域，在字符识别里面ROI继承这个区域就能实现指定位置OCR检测
对于多行OCR，只能单行显示：工业 OCR 工具默认是“单行字符识别”，多行识别就是分开多行识别，这些行如果有空间上的关系，可以用位置调整
歪的有时候检测不到：加位置修正（单次执行后，确定基准），对快速匹配中的设置的特征提取的检测区域进行调整，作为字符检测的位置修正的参数
数据接收和发送：配置VM为服务器，使用NetAssist连接到服务器，然后单次运行程序看是否接收成功
```
**新建流程是在同一文件里面新建的，之前的流程删了就真删了**
