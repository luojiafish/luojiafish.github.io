## 项目2 人脸识别打卡与道具覆盖
思路：
人脸识别打卡
1. 首先定位人脸位置，
2. 然后进行模板匹配，
3. 匹配到就改写打卡txt中每个人的打开状态（默认为NAN）,并显示打卡成功
道具覆盖
1. 使用dlib库：最好python在3.9及以下版本，安装教程如下：https://zhuanlan.zhihu.com/p/627007173
2. 使用detector和perdetector，得到脸的特征点，选择鼻子的或者整个区域选择和一张图片拼接，要下载官方68点的面部识别特征提取文件 https://dlib.net/files/shape_predictor_68_face_landmarks.dat.bz2

```
问题：
只进行模板匹配，不准确 ：尝试先进行人脸识别，给出位置先验后，才进行模板匹配（耗时增加），并没用变得更准确
如果人脸有旋转，畸变等问题：先对模板进行旋转，畸变等操作，然后暴力匹配图像人脸
用conda新建了一个环境，显示无效，原来是之前的环境用conda删了，这里没删，新建的环境没找的，去本地找，教程如下：https://zhuanlan.zhihu.com/p/1916276600011485894
使用detector一直报错RuntimeError: Unsupported image type, must be 8bit gray or RGB image：解决方法调整numpy版本到1.26.4，教程https://blog.csdn.net/weixin_72749746/article/details/140029404，注意cv，dlib之间的兼容版本

```
