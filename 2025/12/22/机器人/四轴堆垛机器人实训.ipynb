# 明确流程
先确定需要用的什么软件和知识，把整个项目的流程摸清楚，在实现过程中学习
1. 下载STM32CubeIDE，跑通项目
2. 阅读电机手册和上位机调试以及单片机调试手册
3. 摸清逻辑
4. 开始单电机，然后多电机调试
5. 逐步实现

## 第一步 跑通项目
问题：汉化不成功，没安装正确的插件，插件之间干扰，下午先看视频学一会
1. 安装 java，安装IDE（编写代码）
2. 安装CubeMX（配置环境生成代码），从 CubeIDE 2.0.0 开始，CubeMX 不再内置在 IDE 里，需要你单独下载安装 STM32CubeMX（独立软件），安装stm32h7的包
3. 在 CubeMX 里打开 .ioc，右上角 Project Manager，Toolchain/IDE：选 STM32CubeIDE，Code Generator 里确认：勾上 “Generate peripheral initialization as a pair of .c/.h files per peripheral”（可选），最重要：勾上 “Keep User Code when re-generating”（保留用户代码），点 Generate Code
4. 然后打开IDE,File → Import… → C/C++ → Existing Code as Makefile Project,Toolchain 选：ARM Cross GCC（一般就这个）,打开文件后点击ioc，然后build
